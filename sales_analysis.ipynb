```python
import pandas as pd
df = pd.read_csv("sales_data.csv")
df
```

```python
print("ðŸ“„ Sales Data:")
df
```

```python
total_revenue = df["Total_Sales"].sum()
```

```python
best_product = df.groupby("Product")["Total_Sales"].sum().idxmax()
best_product_sales = df.groupby("Product")["Total_Sales"].sum().max()
```

```python
print("\nSALES ANALYSIS RESULT")
print(f"Total Revenue: â‚¹{total_revenue}")
print(f"Best Selling Product: {best_product}")
print(f"Revenue from Best Product: â‚¹{best_product_sales}")
```

```python
import  matplotlib.pyplot as plt
plt.figure()
product_total_sales = df.groupby("Product")["Total_Sales"].sum()
product_total_sales.plot(kind="bar")
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales (â‚¹)")
plt.tight_layout()
plt.show()
```
